```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_file = 'Diwali Sales Data.csv'
df = pd.read_csv(csv_file, encoding='latin-1')
```

```python
#Data cleaning process
df.shape
df.head()
```

```python
df.info()
```

```python
#axis means rows and inplace means after dropping save it and drop a blank columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
#last two rows removed
df.info()
```

```python
pd.isnull(df)
```

```python
#check for all null values
pd.isnull(df).sum()
```

```python
df.shape
```

```python
#drop all null values
df.dropna(inplace=True)
```

```python
#reduced 12 rows
df.shape
```

```python
#change datatype
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
#rename column
df.rename(columns={'Marital_Status':'Shaadi'})
```

```python
#describe() return dscription of the data in dataframe
df.describe()
```

```python
#gives min, max, amount of sale, etc.
df[['Amount', 'Orders', 'Age']].describe()
```

```python
# Now starting with Expolatory Data Analysis
```

```python
#gender
df.columns
```

```python
#Return chart of Female order and male order wth the age group
ax = sns.countplot(x='Gender', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#This barplot show female purchase and male purchase
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Gender', y='Amount', data=sales_gen)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
#Age
df.columns
```

```python
#Shows groupwise age order purchased and hue identify men and women
ax = sns.countplot(x='Age Group', hue='Gender', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total amount vs age group here we see most of the buyer are between age 26-35
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

```python
#Total no of orders from top 10 states
sales_states = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

#this give width and height for chart
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State', y='Orders', data=sales_states)
```

```python
#Total amount and sale from top 10 states
#we get to know that most of the orders and total sales/amount is from UP, Maharashtra, Karnataka
sales_amt = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

#this give width and height for chart
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(x='State', y='Amount', data=sales_amt)
```

```python
#Marital Status
ax = sns.countplot(x='Marital_Status', data=df)

sns.set(rc={'figure.figsize':(3,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#most of the buyer are maried women and they have high purchase power
sales_MS = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

#this give width and height for chart
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status', y='Amount', data=sales_MS, hue='Gender')
```

```python
#Occupation
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(x='Occupation', hue='Occupation', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Most of the buyer works in IT, Avaition, Healtcare, etc.
sales_Occupation = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

#this give width and height for chart
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='Occupation', y='Amount', hue='Occupation', data=sales_Occupation)
```

```python
#Product category
sns.set(rc={'figure.figsize':(27,8)})
ax = sns.countplot(x='Product_Category', hue='Product_Category', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Most of the sold product are Food, Footware, etc.
sales_PC = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

#this give width and height for chart
sns.set(rc={'figure.figsize':(20,4)})
sns.barplot(x='Product_Category', y='Amount', hue='Product_Category', data=sales_PC)
```
